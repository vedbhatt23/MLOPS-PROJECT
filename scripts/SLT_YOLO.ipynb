```python
import os
import cv2
from ultralytics import YOLO

# Define the path to the YOLO model weights
model_path = "path_to_your_model_weights.pt"  # Replace with the actual path to your model weights

# Initialize the YOLO model
model = YOLO(model_path)

# Load the image you want to make predictions on
image_path = "path_to_your_image.jpg"  # Replace with the actual path to your image
image = cv2.imread(image_path)

# Perform object detection on the image
results = model(image)

# Access the detection results (bounding boxes, scores, and class IDs)
bounding_boxes = results.xyxy[0].cpu().numpy()  # Extract bounding boxes as NumPy array
class_ids = results.pred[0][:, -1].cpu().numpy().astype(int)  # Extract class IDs as NumPy array
class_names = [results.names[i] for i in class_ids]  # Map class IDs to class names

# Draw bounding boxes and class labels on the image
for box, class_name in zip(bounding_boxes, class_names):
    x1, y1, x2, y2, score = box
    x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])

    # Draw the bounding box
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Add the class label and confidence score
    label = f"{class_name}: {score:.2f}"
    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Display the image with bounding boxes and class labels
cv2.imshow("YOLO Object Detection", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
